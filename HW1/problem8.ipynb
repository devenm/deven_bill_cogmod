```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_points(num_points):
    in_circle = 0
    for n in range(num_points):
        x = np.random.random()*2 # [0 and 2)
        if x > 1:
            x = 1 - x # (-1, 1)
        y = np.random.random()*2
        if y > 1:
            y = 1 - y
        
        if x*x + y*y <= 1:
            in_circle+=1
    return 4 * in_circle/num_points
```

```python
def generate_points_runs(runs, scale):
    approx_error = np.zeros(runs)
    total_points = np.zeros(runs)

    for n in range(runs):
        if n == 0 : continue
        val = generate_points(n*scale)
        approx_error[n] = np.absolute(np.pi-val)
        total_points[n] = n*scale

    return total_points, approx_error
```

```python
runs = 100
scale = 2

total_points, approx_error = generate_points_runs(runs, scale)
# plt.plot(total_points, approx_error)

f, ax = plt.subplots(figsize=(10, 6))
ax.plot(total_points, approx_error)
ax.set_xlim(1-0.001)
ax.set_xlabel('Total Points')
ax.set_ylabel('Approx. Error')
ax.set_title('Pi Estimator Error vs. Points Generated')
ax.grid(alpha=0.3)
```
